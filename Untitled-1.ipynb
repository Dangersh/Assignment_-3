```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
stock_data = pd.read_csv('StockData.csv')
display(stock_data)
```

```python
plt.plot(stock_data)
plt.show()
```

```python
# Plot the data
plt.figure(figsize=(5, 5))
plt.plot(stock_data['Trading Day'], stock_data['AAPL'], label='AAPL Close Price')
plt.xlabel('Trading Day')
plt.ylabel('Close Price')
plt.title('AAPL Stock Close Price Over Time')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(stock_data['Trading Day'], stock_data['AAPL'], label='AAPL Close Price')
ax.plot(stock_data['Trading Day'], stock_data['BNTX'], label='BNTX Close Price')
ax.plot(stock_data['Trading Day'], stock_data['BRK-B'], label='BRK-B Close Price')
ax.plot(stock_data['Trading Day'], stock_data['CSCO'], label='CSCO Close Price')
ax.plot(stock_data['Trading Day'], stock_data['IYK'], label='IYK Close Price')
```

```python
tickers = ['AAPL', 'BNTX', 'BRK-B', 'CSCO', 'IYK']

fig, ax = plt.subplots(figsize=(10, 5))

for I in range(len(tickers)):
    ax.plot(stock_data['Trading Day'], stock_data[tickers[I]], label=tickers[I] + ' Close Price')

ax.set_xlabel('Trading Day')
ax.set_ylabel('Close Price')

ax.set_title('Stock Close Price Over Time')

ax.legend()

plt.show()
```

```python
df = stock_data.loc[:, ~stock_data.columns.isin(['Trading Day'])]

categories = []
average_stock_value = []
for column in df.columns:
    categories.append(column)
    average_stock_value.append(df[column].mean())

fig, ax = plt.subplots(figsize=(10, 5))

ax.bar(categories, average_stock_value)

ax.set_xlabel('Stock')
ax.set_ylabel('Average Close Price')
ax.set_title('Average Stock Close Price')


```

```python
shares = [400,100,200,300,500]

total_portfolio_value = 0.0
portfolio_weights = []

for i,columns in enumerate(df.columns):
    total_portfolio_value += shares[i] * df[columns].tail(1).values[0]
    portfolio_weights.append(shares[i] * df[columns].tail(1).values[0])

portfolio_weights = [portfolio_weights[i] / total_portfolio_value for i in range(len(portfolio_weights))]

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(portfolio_weights, labels=categories, autopct='%1.1f%%')
```
