```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
stock_data = pd.read_csv('StockData.csv')
display(stock_data)
```

```python
plt.plot(stock_data)
plt.show()
```

```python
# Plot the data
plt.figure(figsize=(5, 5))
plt.plot(stock_data['Trading Day'], stock_data['AAPL'], label='AAPL Close Price')
plt.xlabel('Trading Day')
plt.ylabel('Close Price')
plt.title('AAPL Stock Close Price Over Time')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(stock_data['Trading Day'], stock_data['AAPL'], label='AAPL Close Price')
ax.plot(stock_data['Trading Day'], stock_data['BNTX'], label='BNTX Close Price')
ax.plot(stock_data['Trading Day'], stock_data['BRK-B'], label='BRK-B Close Price')
ax.plot(stock_data['Trading Day'], stock_data['CSCO'], label='CSCO Close Price')
ax.plot(stock_data['Trading Day'], stock_data['IYK'], label='IYK Close Price')
```

```python
tickers = ['AAPL', 'BNTX', 'BRK-B', 'CSCO', 'IYK']

fig, ax = plt.subplots(figsize=(10, 5))

for I in range(len(tickers)):
    ax.plot(stock_data['Trading Day'], stock_data[tickers[I]], label=tickers[I] + ' Close Price')

plt.legend()
plt.show()
```
